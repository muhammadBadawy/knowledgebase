# Basic definitions of probability

onsider an experiment whose outcome is not predictable
with certainty in advance.\
However, the set of all possible outcomes is known\
Like rolling a dice or flipping a coin.

**space S** : The set of all possible outcomes of an experiment.\
it's like (1,2,3,4,5,6) for a dice, or (heads, tails) for a fair coin

**Event E** : Any subset E of the sample space S
    - An event is a set consisting of possible outcomes of the experiment (so it's basicly a subset).
    - We can consider the event E occured if the outcome of the experiment is contained inside it (event E).

Example (rolling a die): let E is the event if we get an even
number:

    – If we roll 5 E has not occurred
    – If we roll 2 E has occurred

## What about the occurrence of more than 1 event

**The union ∪** :
For any two events E and F of a sample space S, we define the
new event E**∪**F, called the union of the events E and F, to
consist of all outcomes that are either in E or in F or in both E and F.

So we can say:\
if E =  {2, 3, 4}\
and F = {3, 6}\
then EUF = {2, 3, 4, 6}



**intersection of E and F** :

for any two events Eand F, we may also define the new event E ∩ F, called the intersection of E and F, to consist of all outcomes that are in both E and F. 

so we can say then E∩F = {3}



If there is no intersection between two events Eand G, we get what is called a null event, i.e., an event that has no outcome and will not occur.

So if we say G = {1}\
then E∩G = Φ

** complement $E^c$ **:
the complement of E consists of all outcomes in the sample space S that are not in E, ($E^c$ will ocure if and only if E does not occur)

We can say E $E^c$ = {1, 5, 6}

The properties of the relationships between events:

– Commutative law\
E ∪ F = F ∪ E and E ∩ F = F ∩ E

– Associative law\
(E ∪ F) ∪ G = E ∪ (F ∪ G) and (E ∩ F) ∩ G = E ∩(F ∩ G)

– Distributive law (E ∪ F) ∩ G = (E ∩ G) ∪ (F ∩ G) and (E ∩ F) ∪ G = (E ∪ G) ∩ (F ∪ G)

# Definition of a probability

**Probability or event E  -P(E)-** :
For each event E of an experiment having a sample space S there is a number, denoted by P(E), which is the relative frequency of the event E when a large number of repetitions of the experiment are performed.

From a mathematical view, the probability of the event E, P(E) should satisfy these axioms:

Axioms of Probability:
- Axiom 1 : 0 ≤ P(E) ≤ 1\
- Axiom 2 : P(S) = 1\
- Axiom 3 : For two mutually exclusive events E1 and E2 where E1 ∩E2 = Φ , then P(E1UE2) = P(E1)+P(E2)

Generally, for any sequence of mutually exclusive events E1,E2, . . . (that is, events for which Ei∩Ej =∅ when i=j ),
\


Basic Theorems/Propositions:
- P(Φ) = 0
- P(A') = 1 - P(A)
- If A and B are 2 events such as A is a sunset of B, then P(A) ≤ P(B)
- If A nd B are any 2 events, then P(A-B) = P(A) - P(A∩B)
- If A nd B are any 2 events, then P(A∪B) = P(A) + P(B) - P(A∩B)

```python
# Create function that returns probability percent rounded to one decimal place

def event_probability(event_outcomes, sample_space):
    probability = (event_outcomes / sample_space) * 100
    return round(probability, 1)

# Sample Space
cards = 52

# Determine the probability of drawing a heart
hearts = 13
heart_probability = event_probability(hearts, cards)

# Determine the probability of drawing a face card
face_cards = 12
face_card_probability = event_probability(face_cards, cards)

# Determine the probability of drawing the queen of hearts
queen_of_hearts = 1
queen_of_hearts_probability = event_probability(queen_of_hearts, cards)

# Print each probability
print(str(heart_probability) + '%')
print(str(face_card_probability) + '%')
print(str(queen_of_hearts_probability) + '%')
```

```python
import random
def coin_trial():
    heads = 0 
    for i in range(100):
        if random.random() <= 0.5:
            heads +=1 
    return heads 

def simulate(n):
    trials = []
    for i in range(n):
        trials.append(coin_trial())
    return(sum(trials)/n)

print(simulate(10))
print(simulate(100))
print(simulate(1000))
print(simulate(10000))
```

# Let's Wrap up

We learned to compute probability for:
- a single event from one experiment
- multiple events that are “mutually exclusive” from one experiment. These events cannot happen at the same time.
- Event from an experiment done two or more times (limited number) using counting techniques.

Later you will learn how to compute the probability of events that are not mutually exclusive (jointly occur) and their outcomes might mutually depend (conditional occurrence).

Later you will learn how to compute the probability of events/outcome of an experiment repeated for a large number of times where the space is not equi-probable (probability density)
