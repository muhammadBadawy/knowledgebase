Suppose that in order to test our new drug, our null hypothesis is that the effect of the new drug is the same as the effect of the standard treatment.

The alternative hypothesis is that they are not the same.







In this case, what I just described is called a two sided test, because there are two ways in which

the drug can be different from the control.

The effect of the drug can be less or the effect of the drug can be greater.

Those are two different scenarios under the same hypothesis, on the other hand.

The alternative hypothesis may simply be that the effect of a drug is greater than the effect of the

standard treatment.

In this case, we call it a one sided test.

So the notation for this is slightly different since the alternative now is that mu is greater than the one we use a greater than sign instead of a not equal sign.

Another way to think of this is that the distribution of Group two could be to the left or to the right of Group one.



Note that the same concepts can apply to our stock return example, which is a one sample test we can check if the mean daily stock return is not equal to zero, or we can check if the mean daily stock return is greater than zero, specifically, since it being less than zero is not necessarily something

we care about.

Again, these are two sided and a one sided test respectively.

OK, so these are just two ways we can categorize different kinds of tests.

We can have one sample test or two sample tests.

We can also have one sided tests or two sided tests.

So what have we learned?

We've learned that comparing two groups is not simply just a matter of comparing their averages, their averages will always be different.

In every experiment, one must have a higher average than the other.

So this is not a useful thing to compare.

Furthermore, we have uncovered two crucial new variables that do seem to be useful.

They are, number one, the number of samples collected and number two, the variance in the samples.

This should remind you exactly of confidence intervals, since these two items are also what affects the width of the confidence interval.

Generally speaking, all of these variables are highly relevant when we assess whether or not the results of our hypothesis test is statistically significant.

If the difference between the average of the two groups is not large enough, then the result is not statistically significant.

- If there are not enough samples, then the result is not statistically significant.
- If the variance in the data is too large, then again, the result is not statistically significant.

We will learn how to quantify statistical significance.
