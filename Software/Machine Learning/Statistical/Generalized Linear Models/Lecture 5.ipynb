alpha could be positive or negative (add or subtract some of b to a) based on the need to vector a to be more like b or less like b

Back to our rule we made (used in GD, logistic regression, and linear regression), we will notice that I'm adding some (alpha) of x to theta to make the new theta closer to x

The part (y(i) - the hypothesis) defined whether I'm going to add or subtract some of x from X,
and don't mix the alphas, because alpha here is the learning rate, not the one in the lower left corner examples.

the key takeaway from the perceptron algorithm is this idea that if you have two vectors and their dot product is smaller than what you would desire it to be, then take a small amount of one of the vectors and add it to the other, And that will make the dot product bigger.

And similarly, if the dot product is larger than what you would like it to be, then take a small amount of one vector and subtract it from the other vector.

(remember that the dot product is called the projection product)\
(the dot product is just an unnormalized version cos the angle between the 2 vectors x and theta)

# Comparison between the perceptron and logistic regression

Perceptron
- The output of h is 0 or 1 (right or wrong)
- Only use update if we got the example wrong, because in right examples the error is just 0, and this way there will be no need to take part of x and add it to theta
- The update is just taking a part ($\alpha$) from the data example x and (adding to or subtracting from) $\theta$
- The $\theta$ is a linear combination of the misclassified data examples

Logistic Regression
- the output of h is never 0 or 1 (something between them)
- Updates on every training example (because of there will be always some error duo to the output that will never hit 0 or 1 -unlike the perceptron that gets 0 or 1 always, so some cases the error is just 0, and this way there will be no need to take part of x and add it to theta-), so it always takes a piece of x depending on the degree of how much we got it right.
- The $\theta$ is a linear combination of the all data examples
- Soft version of perceptron
- uses a threshold
- the output is a probability
- you can sample your output based on the probability you got

Newton's method is a root finding method, and when we are trying to find the root at the first derivative of a function (the point 0 in the first derivative is a maxima or minima) we may find the root of a local or global maximum point

Newton's method will take you to the closest stationary point (minima, maxima, or saddle point), unlike the GD that will take you to the always to the minima.\
Newton's method is pretty ok when we are dealing when we are close to the global minimum, so it will be good with convex function where the nearest stationary point is the global minimum

so you only want to use is with concave (convex) functions

Newton's method doesn't work well with SGD as it requires $H^{-1}$ to be invertable, and with low data points H can't be inverted
