OK, so this is a very intuitive idea behind the confidence interval, what we're trying to do is we're

trying to create a distribution of the sample mean or any other quantity.

We are trying to estimate the confidence interval is just an interval over this distribution.

For example, you can choose the middle 95 percent.

This is called the ninety five percent confidence interval.

And from this intuitive example, we see that when the variance of our original data is larger, this makes our confidence interval larger.
(the smaller the interval, the more we re sure about the actual mean -not the sample mean- we got from expecting the sample mean)

However, we can shrink our confidence interval by collecting more data.

That is, we can become more confident in our estimate by collecting more samples.

OK, so more variance means larger and more samples mean smaller.
