# Decomposition

we have 3 types:
- Singular Value Decomposition
- EigenValue Decomposition
- Cholesky Decomposition



For SVD:
- U and V are orthonormal unrelated matrices
- S the values at the diagonal are always singular values

For EVD:
- U is a matrix of the eigenvectors as columns
- D the values at the diagonal are the eigenvalues
- Only valid for square matrices, because eigenvectors are only in square matrices

So for EVD, the behavior of the transformation matrix is defined by the eigenvectors U and the eigenvalues D



The matrix will be split into 3 matrices that will act like a pipeline

So we can review the steps of the decomposition



So we are
- rotating , at EVD u r are sending the vectors to eigenvector basis world
- then scaling
- undo the rotation, at EVD, this rotation is the opposite of the first rotation

At step 2 generally, after the first transformation when we are multiplying with diagonal matrix it's just pure scaling along the axis by the component in the diagonal entry.

At step 2 for EVD, after the first rotation you will see that you your axis align with your eigenvectors (your axis become the eigenvectors -you moved to eigenvectors basis world-)\
So whatever rotation (transformation) you will do after the first rotation, it will not affect the vector direction after the second rotation (undoing the first rotation -returning to old basis world, or we could say that the axis align with it's old location-)

So SVD deals with any matrix and provides S the diagonal matrix with singular values.\
And EVD deals with only square matrices and provides D the diagonal matrix with eigenvalues.

SVD for Any matrix\
EVD for Any square matrix\
EVD and SVD are the same for symmetric square matrices\
