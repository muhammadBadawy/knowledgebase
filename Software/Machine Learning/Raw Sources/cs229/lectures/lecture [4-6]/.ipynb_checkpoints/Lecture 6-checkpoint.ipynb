Base measure : which is typically any distribution of the data you got (exponential or non exponential) distribution.

So The process that we are doing is converting the base measure b(y) to become exponential exponential, and this process is called **exponential tilting** (and the result here is another measure).\
so basically we drove a new measure (new distribution) out of the base measure (old distribution) by doing exponential tilting.

after the exponential tilting, we get a normalized distribution that integrates/sums to one (a normalized version of the base measure)\
Example: converting raw data to normal distribution (note here that the integration of area of the ND is one)

So in total this is what happens



- First we get our data sample $X^{(i)}$
- We multiply this x with the learned $\theta$ -> now you got $\eta$
- Now calculate the 3 components T(y), a($\eta$), and b(y)
- Now choose what exponential distribution you want to sample your output y from (this is the whole idea for this lesson, choose the distribution that fits your output the best)
- Now we got our pipeline that we feed $X^{(i)}$ to it to get the probability distribution of y
- We use this probability distribution of y to get expected value of y E[y|x]
- The expected value is the final prediction

And the case for the logistic regression is that.\
Let's walk the process of doing it the naive way, then redoing it using GLM ways to get the privileges of the GLM.

If we think of it independently **without looking at the GLM**, we will think as follow:

I have the linear regression model and we know why we can't use it to solve a logistic regression.\
So we want to produce output 0 or 1, so we may choose to get a function that surrounds the linearity of the linear regression $g(Z)$ = $g(\theta^TX)$\
So naively we can choose a function that set the output y to 1 if Z >= 0 and to 0 if Z < 0 (like the perceptron)

So the result is the perceptron algorithm, and now we don't have the properties of the GLM

So what if we want to get these nice properties of the GLM?

- The answer is easy, we can think of the output y as a probability distribution, and derive it from a GLM construction process.
- All the GLMs has the same update rule

So we need to ask ourselves what is the distribution that gives a similar output to the one we want, and belongs to the exponential family at the same time?\
The answer is easy, it is the Bernoulli distribution.

Note:\
Never forget the three assumptions we made before building a GLM (specially the third one $\eta = \theta^TX$)

- so First we get our data sample $X^{(i)}$
- We multiply this x with 𝜃 -> now you got 𝜂 (third assumption)
- Now calculate the 3 components T(y), a(𝜂), and b(y)
- So we chose the Bernoulli distribution, let's build it using the components we got
- Now we got out pipeline that we feed $X^{(i)}$ to it to get the probability distribution of y
- We use this probability distribution of y to get expected value of y E[y|x] and it is for sure the mean $\phi$ ($\phi$ is the name of the mean of the Bernoulli)
- now we use the form of the mean where it is a function of the natural parameter $\eta$, and it is $\phi = \frac{1}{1 + e^{-\eta}}$ -> and this is the logistic function (now we know why we did choose this function over we chose at the perceptron algorithm, the fun thing that we didn't choose it, it did came naturally after the GLM construction process with the Bernoulli)
- The expected value $\phi$ (the mean) is the final prediction

Softmax Classifier


Friendly reminder:



Here we have:\
data in 2 dimensions and the third dimension is embedded to the shape of the data point

The $\theta$ is a vector where it gives high positive value, and when it is dotted with other group of the data, it gives very negative value.

Here you see that the separating hyperplane is the line perpendicular to learned parameters $\theta$, why? because if we dot product that line it is undefined, and every data point of that line doesn't belong to any class.

so the side that looks towards the class of the $\theta$ is where every point gives positive value ($\theta^TX$) when it's dotted with the learned parameter $\theta$

and the side that looks towards the opposite of the $\theta$ is where every point gives negative value ($\theta^TX$) when it's dotted with the learned parameter $\theta$

--

And the same goes for the softmax classifier where we have multiple classes




So when we want to classify a point, we dot every $\theta$ and then argmax() them



So the result looks like this 



we need to get rid of the negative values without losing the the relative order (we have seen similar cases in linear algebra, and it's literally like converting into another dimension without loosing any data), so we will use the exp on the result.



And then normalize it\
we just summed (integrated them if continues data) them and divided them by the total



(And this process is called softmax function)



So someone may raise a question, why can't just we just get the normal max (argmax) of all ($\theta^TX$) instead of the softmax?



And the answer is that the softmax function has some good properties such as
- Differentiable
- Can do Gradient descent

it's just a differentiable, smoother way to get the max
